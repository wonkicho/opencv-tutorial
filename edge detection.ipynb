```
#canny edge detection

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```
img = cv2.imread('/content/drive/My Drive/vision/DATA/sammy_face.jpg')
```

```
plt.imshow(img)
```

```
edges = cv2.Canny(img, threshold1=127, threshold2=127)
```

```
plt.imshow(edges,cmap='gray')
```

```
edges = cv2.Canny(img, threshold1=0, threshold2=255)
plt.imshow(edges,cmap='gray')
```

```
#calculate median pixel
med_val = np.median(img)
```

```
med_val
```

```
#lower threshold to either 0 or 70% of the median value
lower = int(max(0,0.7*med_val))
#upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255,1.3*med_val))
```

```
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper+100)
plt.imshow(edges)
```

```
blurred_img = cv2.blur(img, ksize=(7,7))
edges= cv2.Canny(image=blurred_img, threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```
lena = cv2.imread('/content/drive/My Drive/vision/DATA/lena.jpg')
lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)
plt.imshow(lena)
```

```
med = np.median(lena)
lower = int(max(0,0.7*med))
upper = int(max(255,med*1.3))

blur_lena = cv2.blur(lena,(2,2))
edges = cv2.Canny(blur_lena,threshold1=lower, threshold2=upper)
```

```
plt.imshow(edges,cmap='gray')
```
