```
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#imread의 두번째 파라메터 0을 설정함으로써, grey scale설정
img = cv2.imread('/content/drive/My Drive/vision/DATA/rainbow.jpg',0)
plt.imshow(img, cmap='gray')
```

```
#255/2 -> 127
#source, threshold, max value
#thresh_binary로 0 아니면 1 -> black or white, inv붙이면 그 반대로
#threshold trunc -> 임계값보다 높으면 임계값으로, 아니면  0
ret, thresh1=cv2.threshold(img,127,255, cv2.THRESH_BINARY)
```

```
#임계값보다 높으면 1 낮으면 0
plt.imshow(thresh1,cmap='gray')
```

```
img = cv2.imread('/content/drive/My Drive/vision/DATA/crossword.jpg',0)
plt.imshow(img, cmap='gray')
```

```
def show_pic(img):
  fig = plt.figure(figsize=(15,15))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap='gray')

show_pic(img)
```

```
ret, th1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```
blended=cv2.addWeighted(src1=th1,alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
